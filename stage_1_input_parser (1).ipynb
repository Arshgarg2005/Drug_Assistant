```python
from cosyvoice.cli.cosyvoice import AutoModel

cosyvoice = AutoModel(
    model_dir="/content/pretrained_models/CosyVoice2-0.5B"
)
```

```python
import librosa
from pathlib import Path
import numpy as np

# Find speaker reference file
speaker_ref_path = "/content/speaker_references/new.wav"

# Load as 16kHz (required by CosyVoice)
print(f"Loading speaker reference from: {speaker_ref_path}")
speaker_ref_16k, sr = librosa.load(speaker_ref_path, sr=16000)

print(f"✓ Loaded: {len(speaker_ref_16k)/16000:.1f}s at {sr} Hz")
print(f"  Shape: {speaker_ref_16k.shape}")
```

```python
import torchaudio
import numpy as np

# Test text
test_text = "What were the biggest challenges you faced during your recovery journey? Can you elaborate on at least two?"

# Prompt text must match what's in the speaker reference audio
# (Or use empty string if you just want voice cloning without matching)
prompt_text = "are the biggest up's and down's that have come in your journey to recovery"  # Empty for zero-shot without exact text matching

print("Synthesizing audio...")
print(f"  Text: {test_text}")
print(f"  Speaker ref duration: {len(speaker_ref_16k)/16000:.1f}s")

# IMPORTANT: CosyVoice2 returns generator, iterate through results
audio_data = None

for i, result in enumerate(
    cosyvoice.inference_zero_shot(
        tts_text=test_text,
        prompt_text=prompt_text,
        prompt_wav="/content/speaker_references/new.wav",   # <-- changed
        stream=False
    )
):
    audio_data = result["tts_speech"]
    print(f"  Generated: {audio_data.shape} at {cosyvoice.sample_rate} Hz")

if audio_data is not None:
    print("✓ Synthesis successful!")

    # Save
    output_path = "/content/output/test_synthesis.wav"
    torchaudio.save(output_path, audio_data, cosyvoice.sample_rate)
    print(f"✓ Saved to: {output_path}")
else:
    print("✗ No audio generated")
```
